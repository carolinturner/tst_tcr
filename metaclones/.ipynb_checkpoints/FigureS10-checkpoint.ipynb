```python
import os
import time
import pandas as pd
from os.path import join, dirname, abspath
import contextlib

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl


import igraph
import seaborn as sns
import scipy.stats

import pyrepseq as prs
import pyrepseq.plotting as pp
from metaclonotypist import *

plt.style.use('seaborn-v0_8-paper')
```

# Load and filter HLA data

```python
hlas = pd.read_csv(f'data/hladata.csv', index_col=0)
hlas = flatten_hlas(hlas)
hlas.head()
```

```python
testmethod = 'fisher'
min_donors = 4
```

```python
# filter hlas < min_donors
hlas = hlas[hlas.columns[hlas.sum(axis=0)>=min_donors]]
# filter class I HLAS
hlas = hlas[hlas.columns[hlas.columns.str.startswith('D')]]
len(hlas.columns)
```

# Load and filter Gliph2 output

```python
clusters = pd.read_csv(f'data/gliph2_beta_mhcII_cluster.csv')
len(clusters['index'].unique())
```

```python
clusters = clusters[clusters['number_subject']>=4]
len(clusters['index'].unique())
```

```python
clusters = clusters[clusters[' number_unique_cdr3']>=3]
len(clusters['index'].unique())
```

```python
clusters = clusters[clusters['vb_score']<0.05]
len(clusters['index'].unique())
```

```python
clusters = clusters[clusters['length_score']<0.05]
len(clusters['index'].unique())
```

```python
clusters = clusters[clusters['Fisher_score']<0.05]
len(clusters['index'].unique())
```

```python
clusters.to_csv('data/gliph2_filter1.csv')
```

# Perform HLA association analysis

```python
clusters_rn = clusters.rename(columns=dict(index='cluster', Sample='Sample.ID'))
```

```python
cluster_association = hla_association(clusters_rn, hlas,
                                      method=testmethod)
```

```python
cluster_association['CDR3s'] = cluster_association['cluster'].apply(lambda x:
                                    '|'.join(clusters_rn[(clusters_rn['cluster']==x)][f'TcRb']))
```

```python
cluster_association.to_csv('data/gliph2_filter1+2.csv')
```
